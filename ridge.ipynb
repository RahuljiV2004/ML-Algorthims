RIDGE REGRESSION

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
x=df.drop('sales',axis=1)
```

```python
y=df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pol=PolynomialFeatures(degree=3,include_bias=False)
```

FEATURE SCALING

```python
pol1=pol.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
pol1.shape
```

```python
xt,xte,yt,yte=train_test_split(pol1,y,test_size=0.3,random_state=41)
```

```python
help(train_test_split)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(xt)
```

```python
scaled_xt=scaler.transform(xt)
```

```python
scaled_xte=scaler.transform(xte)
```

```python
scaled_xt.shape
```

```python
scaled_xte.shape
```

```python
from sklearn.linear_model import Ridge
```

```python
help(Ridge)
```

```python
ridge_model=Ridge(alpha=10)
```

```python
ridge_model.fit(scaled_xt,yt)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
ri=ridge_model.predict(scaled_xte)
```

```python
ri
```

```python
mean_absolute_error(yte,ri)
```

```python
mean_squared_error(yte,ri)
```

```python
root_mean_squared_error(yte,ri)
```

```python
from sklearn.linear_model import RidgeCV
```

```python
RidgeC=RidgeCV([10.0,0.1,13.0],scoring="neg_mean_absolute_error")
```

```python
RidgeC.fit(scaled_xt,yt)
```

```python
RidgeC.alpha_
```

```python
x12=RidgeC.predict(scaled_xte)
```

```python
x12
```

```python
mean_absolute_error(yte,x12)
```

```python
mean_squared_error(yte,x12)
```

```python
root_mean_squared_error(yte,x12)
```
